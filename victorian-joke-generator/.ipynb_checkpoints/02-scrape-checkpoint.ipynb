```python
import requests

url = 'http://victorianhumour.com/jokedb/?offset=40&paging=40'
headers = {'user-agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 9_2 like Mac OS X)'}
response = requests.get(url, headers=headers)

# We are testing for the code 200
response.status_code
```

```python
# We don't need the rest of the bs4 package
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, 'lxml')

soup.prettify
```

```python
soup.body.j
```

```python
j_array = soup.body.find_all('j')

j_array[3].text
```

```python
# for i in j_array:
#     print(i.text)
```

```python
with open('xyz.txt', 'w') as myfile:
    for line in j_array:
        myfile.write("%s\n" % line.text)
```
